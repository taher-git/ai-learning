```python
# Download Titanic dataset from Kaggle and load it into a DataFrame
titanic_df = pd.read_csv('titanic/train.csv')
# print(titanic_df.head())
test_df = pd.read_csv("titanic/test.csv")
# print(test_df.head(10))
# Count missing values per column.
missing_values = test_df.isnull().sum()
print("Missing values per column:")
print(missing_values)
# Fill missing values with mean/median.
test_df['Age'].fillna(test_df['Age'].mean(), inplace=True)



print(test_df.head(50))
```

```python
female_over_30 = test_df[(test_df['Sex'] == 'female') & (test_df['Age'] > 30)]
# Select only Name and Age columns
female_over_30 = female_over_30[['Name', 'Age']]

print(female_over_30)
```
