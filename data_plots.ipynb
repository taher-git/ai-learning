```python
import pandas as pd

df = pd.read_csv('titanic/train.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

survival_counts = df['Survived'].value_counts()

plt.bar(survival_counts.index, survival_counts.values, color=["red", "green"])
plt.xticks([0, 1], ["Did not survive", "Survived"])
plt.ylabel("Count")
plt.title("Survival Counts")
plt.show()
```

```python
import seaborn as sns

sns.countplot(data=df, x="Survived", palette="Set1")
plt.title("Survival Counts")
plt.show()
```

```python
sns.histplot(data=df, x="Age", bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(data=df, x="Pclass", y="Age", palette="Set3")
plt.title("Age distribution by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.savefig("my_correlation_plot.png", dpi=300)

plt.show()
```

```python
x_values = [1, 2, 3, 4, 5]
y_values = [2, 5, 3, 8, 4]

# Create the plot using plt.plot()
plt.plot(x_values, y_values)
plt.plot(x_values, y_values, label="Linear Trend")  # Use the label parameter for the legend
plt.title("My First Matplotlib Plot")
plt.xlabel("Pollulation in millions")
plt.ylabel("Rate of employment")
plt.legend()  # Display the legend based on the 'label' in plt.plot()
plt.savefig("my_plot.png", dpi=300)
plt.show()
```
